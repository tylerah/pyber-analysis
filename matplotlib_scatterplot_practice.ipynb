```python
# dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create scatter plot
plt.scatter(y_axis, x_axis, color = "red", label = "Chicago")
plt.legend()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.gca().invert_yaxis()
plt.xlim(0, 50)
```

```python
# create scatter plot with bubbles
plt.scatter(y_axis, x_axis, s = [i * 3 for i in y_axis], color = "red", label = "Chicago")
plt.legend()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.gca().invert_yaxis()
plt.xlim(0, 50)
```

```python
# create scatter using object oriented approach
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, color="skyblue", label="Chicago", linewidths=2, edgecolors = "black", s = [i * 3 for i in y_axis])
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.legend()
ax.set_title("PyBer Fare by Month")
```
